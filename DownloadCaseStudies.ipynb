```python
import json
import requests

import time

def grab_list_of_ids():
    # use this list of caseStudy Id's
    text_file = open("listOfIds", "r")
    ids = text_file.read().splitlines()
    text_file.close()
    print ("total Number of Studies:", len(ids))
    return ids

def batch_the_ids(ids):
    batchedIds = []
    
    start = 0;
    end = 200;
    while(start < len(ids)):
        split = ids[start:end]
        batchedIds.append(','.join(split))
        start = end
        end = end + 200;
    
    return batchedIds


def make_batch_api_call(id_list):
    response = requests.get("http://impact.ref.ac.uk/casestudiesapi/REFAPI.svc/SearchCaseStudies?ID=" + id_list)
    data = response.json()
    return data

# A place to save all the retieved data
allCaseStudies = []

# Load the list of all Id's
ids = grab_list_of_ids()
# Batch these to make more efficent use of the API
batchedIds = batch_the_ids(ids)

# Fire an API call for each batch
for batch in batchedIds:
    allCaseStudies += make_batch_api_call(batch)
    time.sleep(5)
    
for study in allCaseStudies:
    print(study["Title"])
    
```

```python
for study in allCaseStudies:
    print(study["CaseStudyId"])
```

```python
with open('allStudies.json', 'w') as outfile:
    json.dump(allCaseStudies, outfile)
```
